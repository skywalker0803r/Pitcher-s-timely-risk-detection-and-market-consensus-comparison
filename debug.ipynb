```python
import os
import cv2

def count_video_frames_in_folder(folder_path):
    F = []
    # æœå°‹æ‰€æœ‰ .mp4 æª”æ¡ˆ
    video_files = [f for f in os.listdir(folder_path) if f.lower().endswith(".mp4")]

    if not video_files:
        print("âš ï¸ è³‡æ–™å¤¾å…§æ²’æœ‰ .mp4 æª”æ¡ˆ")
        return

    for video_file in video_files:
        video_path = os.path.join(folder_path, video_file)
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"âŒ ç„¡æ³•é–‹å•Ÿå½±ç‰‡ï¼š{video_file}")
            continue

        # å–å¾—ç¸½å¹€æ•¸
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        print(f"ğŸ¬ {video_file}ï¼š{total_frames} å¹€")
        cap.release()
        F.append(total_frames)
    return F
```

```python
import cv2
import os
import mediapipe as mp

# åˆå§‹åŒ– MediaPipe Pose
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

def process_and_display_videos(folder_path):
    # æ‰¾å‡ºæ‰€æœ‰ mp4 æª”æ¡ˆ
    video_files = [f for f in os.listdir(folder_path) if f.lower().endswith(".mp4")]

    if not video_files:
        print("âš ï¸ è³‡æ–™å¤¾å…§æ²’æœ‰ mp4 æª”æ¡ˆ")
        return

    pose = mp_pose.Pose(static_image_mode=False, # ä½¿ç”¨å½±ç‰‡æ¨¡å¼ï¼Œè‡ªå‹•è¿½è¹¤é—œéµé»
                        min_detection_confidence=0.5, # é è¨­å€¼ï¼Œå¹³è¡¡æº–ç¢ºåº¦èˆ‡è¦†è“‹ç‡
                        smooth_landmarks=True, # å•Ÿç”¨å¹³æ»‘ï¼Œæ¸›å°‘ jitter
                        )

    for video_file in video_files:
        video_path = os.path.join(folder_path, video_file)
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"âŒ ç„¡æ³•é–‹å•Ÿå½±ç‰‡ï¼š{video_file}")
            continue

        print(f"â–¶ï¸ æ­£åœ¨æ’­æ”¾ï¼š{video_file}")

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = pose.process(image_rgb)

            if results.pose_landmarks:
                mp_drawing.draw_landmarks(
                    frame, 
                    results.pose_landmarks, 
                    mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(0,255,0), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(0,0,255), thickness=2)
                )

            cv2.imshow("Pose Detection", frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                print("â¹ ä¸­æ­¢æ’­æ”¾")
                cap.release()
                cv2.destroyAllWindows()
                pose.close()
                return

        cap.release()

    cv2.destroyAllWindows()
    pose.close()
    print("âœ… æ‰€æœ‰å½±ç‰‡æ’­æ”¾å®Œç•¢")


process_and_display_videos('data/Yu_Darvish_FF_videos_4S')
```
